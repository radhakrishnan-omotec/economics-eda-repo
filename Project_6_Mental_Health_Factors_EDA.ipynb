```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Read the CSV file
df = pd.read_csv("/content/economics-eda-repo/Output-Health-Economics-EDA-CSV.csv")

# Check the first few rows of the dataframe
print(df.head())

# Check the columns in the dataframe
print(df.columns)

# Assuming "Q1_new" column exists, generate pie chart for its value counts
if 'Q1_new' in df.columns:
    # Count the occurrences of each value in the 'Q1_new' column
    q1_counts = df['Q1_new'].value_counts()

    # Plotting the pie chart
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))



    # Pie chart
    wedges, labels, autopct = ax[0].pie(q1_counts, labels=q1_counts.index, autopct='%1.1f%%', startangle=140)
    ax[0].legend(wedges, q1_counts.index, title="TITLE", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))
    custom_labels = [str(i) for i in range(6)]
    plt.gca().set_prop_cycle(None)  # Reset color cycle
    for i, custom_label in enumerate(custom_labels):
        plt.plot([], label=custom_label)
    ax[0].set_title('TITLE')
    ax[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.



    # Count plot
    top_four_categories = q1_counts.nlargest(4).index
    sns.countplot(x='Q1_new', data=df[df['Q1_new'].isin(top_four_categories)], ax=ax[1], order=top_four_categories, palette='viridis')
    ax[1].set_title('Count of Top Three TITLE')

    plt.tight_layout()
    plt.show()
else:
    print("Column 'Q1_new' not found in the CSV file.")
```
